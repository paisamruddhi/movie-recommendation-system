```python
import pandas as pd
"""
For the movie recommendation system we are using two files movies.csv and ratings.csv we are using pandas to perform necessary operations on data frame
"""
movies=pd.read_csv('movies.csv')
ratings=pd.read_csv("ratings.csv")
```

```python
"""
next step is to clean the data remove unnecessary data and modify based on our requirements
"""
movies.drop('genres',axis=1,inplace=True)#removes the genre column since we are recommending based on rating here
movies
```

```python
"""
The below lines are to modify the title column such that it would be close to match the user input
"""
movies['title']=movies['title'].str.replace(r"\(.*?\)","")#to remove year ex:(2014)
movies['title']=movies['title'].str.replace("'","")#to remove single quotes example '71 
movies['title']=movies['title'].str.replace(', The','')# removs The at the end example:Shawshank Redemption, The
movies['title']=movies['title'].str.strip()#to remove wide spaces at the end
movies['title']
```

```python

"""
now we merge both the dataframes 'movies' and 'ratings'
"""
dataset = pd.merge(ratings,movies,on='movieId')#movieId is common in both 'movies' and 'ratings' so we merge both dataframes based on 'movieId'
dataset
```

```python
"""
next step is to find mean rating of the movie and the number of user ratings
"""
dataset_based_on_ratings = pd.DataFrame(dataset.groupby('title')['rating'].mean())
dataset_based_on_ratings['number of ratings']=pd.DataFrame(dataset.groupby('title')['rating'].count())
dataset_based_on_ratings
```

```python
"""
Next step is to create a matrix that relates userId and movies based on mean ratings of the movie
"""
movie_matrix = dataset.pivot_table(index='userId',columns='title',values='rating')
movie_matrix
```

```python
x=input("Enter a movie Name ").split()
temp=""
for i in x:
   temp=temp+i.capitalize()+" "
x=temp
movie_watched = movie_matrix[x.strip()]
print("loading.....")
li = []
for i in range(len(movie_matrix.columns)):
    li.append(movie_watched.corr(movie_matrix.iloc[:,i]))
li = pd.Series(li)
df = pd.DataFrame({"title": movie_matrix.columns,"Correlation": li,"number of ratings" : dataset_based_on_ratings["number of ratings"].values})
df[df["number of ratings"] >= 100].sort_values(by=["Correlation","number of ratings"],ascending=False).iloc[1:11,:]

```
